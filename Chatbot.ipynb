```python
# Simple Rule-Based Chatbot

def chatbot():
    print(" ChatBot: Hello! I am a simple chatbot.")
    print("Type 'exit' anytime to end the chat.\n")

    while True:
        user_input = input("You: ").lower()

        if user_input == "exit":
            print(" ChatBot: Goodbye! Have a great day!")
            break

        elif user_input in ["hi", "hello", "hey"]:
            print(" ChatBot: Hello there!")

        elif user_input in ["how are you", "how are you?"]:
            print(" ChatBot: I'm just a program, but I'm doing great!")

        elif user_input in ["what is your name", "your name"]:
            print(" ChatBot: I am a rule-based chatbot.")

        elif user_input in ["help"]:
            print(" ChatBot: You can greet me, ask my name, or type exit.")

        else:
            print(" ChatBot: Sorry, I don't understand that.")

if __name__ == "__main__":
    chatbot()
```
